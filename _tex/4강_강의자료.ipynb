## 조건문의 기본 형태

  - if … :
    - 실행될 코드
  - else:
    - 실행될 코드



```
if 평가할 조건: # 반드시 :(콜론)으로 끝나야 함
  조건이 참일 경우 실행될 코드(1) # 반드시 들여쓰기(indent)가 되어야 함. 여기서부터 들여쓰기가 끝나는 부분까지 순차적으로 실행
  조건이 참일 경우 실행될 코드(2) # 들여쓰기는 보통 키보드의 탭(tab)버튼으로 입력
else: # 반드시 :(콜론)으로 끝나야 함
  조건이 참이 아닐 경우 실행될 코드(1)
  조건이 참이 아닐 경우 실행될 코드(2)
```



````python
# 조건문 실행 예시

a = 2 #a 변수에 숫자형 값 2 입력

if a > 5: # 만약 a 변수 안에 있는 숫자형 값이 5보다 크다면
  print("a는") # 이 문구를 결과창에 출력하고
  print("5보다") # 그 다음 이 문구를 출력하고
  print("크다") # 그 다음 이 문구를 출력한다.
else: # if에 있는 조건이 달성되지 않는다면(= a 변수 안에 있는 숫자형 값이 5보다 크지 않다면)
  print("a는") # 이 문구를 결과창에 출력해라
  print("5보다") # 그 다음 이 문구를 출력하고
  print("크지 않다") # 그 다음 이 문구를 출력한다.

# 만약 인덴트가 맞지 않는다면?
# 11번째 줄 코드를 앞으로 땡겨보자. 뒤로 밀어보기도 하자
````

### 비교 연산자

- 조건문에서 가장 중요한 부분은 바로 ‘조건’
- 이 조건을 구축하는데 사용되는 것이 비교연산자
  - 입력된 두 변수나 값을 비교하여 참인지 거짓인지의 결과를 알려주는 연산자
  - 꼭 조건문에서만 사용되는 것은 아님
  - 비교연산자를 사용한 결과가 바로 bool

- 비교연산자의 종류
  - 숫자형 비교

```
x < y: x가 y보다 작으면 True, 아니면 False
x <= y: x가 y보다 작거나 같으면 True, 아니면 False
x > y: x가 y보다 크면 True, 아니면 False
x >= y: x가 y보다 크거나 같으면 True, 아니면 False
x == y: x와 y가 같으면 True, 아니면 False
x != y: x가 y가 다르면 True, 아니면 False
```




````python
# 텍스트 자료형에서는?

x = "Lee"
y = "lee"

print(x == y)
print(x != y)
print(x < y) #??? 왜?

# 문자를 기계어로 인식하게 해주는 아스키 코드
# 아스키 코드의 숫자를 기반으로 생각하기 때문에 문자형 비교도 작동
# 원하는 바가 아니기 때문에 문자형의 경우 이런 연산기호로 비교하지 않음
````

  - 문자형 비교

```
==, != : 좌변 우변 문자가 동일한가
in, not in : 좌변에 있는 문자가 우변에 있는 문자에 포함되어 있는가
B.startswith(A) : B 문자형이 A 문자형으로 시작하는가
B.endswith(A) : B 문자형이 A 문자형으로 끝나는가
```



````python
print('a' in 'apple')
print('a' not in 'apple')
print('k' not in 'apple')
print('apple'.startswith('app'))
print('apple'.startswith('ppl'))
print('apple'.endswith('app'))
print('apple'.endswith('ple'))
````

### 논리연산자

- 논리연산자는 논리식을 두고 조건에 따라 참인지 거짓인지를 판단
  - 주로 and, or, not이 사용됨

```
x and y: x와 y 모두 True여야 True 반환
x or y: x와 y 둘 중 하나만 True여도 True 반환
not x: x가 거짓이면 True 반환
x: x가 참이면 True 반환
```



````python
x = 3
y = 4

print(x > y) # x가 y보다 크면 True, 아니면 False
print(x != y) # x와 y가 다르면 True, 아니면 False

print(x > y and x != y) # False and True. 그러니까 False
print(x < y and x != y) # True and True. 그러니까 True
````

# 반복문

## while문

```
while 조건:
  반복할 내용
```
- 조건이 핵심. 조건이 참인 상황에서 반복할 내용을 무한하게 반복할 수 있음
- 어떤 상황이 되면 내가 조건을 False로 만들어 반복을 종료할 것인가의 관점이 중요


````python
# while문 기초

# import time

a = 0 # a에 값을 입력
# time.sleep(3)
while a < 5: # a가 5보다 작을 때
  # time.sleep(3)
  print(f"a = {a}") # 이 문구를 인쇄하고
  # time.sleep(3)
  a = a + 1 # a에 들어있는 값에 1을 더해서 a에 저장
  # time.sleep(3)

print('end')
# time.sleep(3)

# 실행 순서는?
# 5, 7, 9, 11, 7, 9, 11, 7, 9, 11, 7, 9, 11, 7, 9, 11, 14
````

## list comprehension

반복문을 사용해 쉽게 리스트를 만드는 방법

쉽게? 여러 라인의 코드를 한 줄로 표기한다는 것. easy의 개념과는 약간 다름

````python
# 리스트의 각 요소에 3을 곱하여 다시 리스트 만들기

a = [1,2,3,4]
result = []
for num in a:
  result.append(num*3)

print(result)
````

````python
# list comprehension 사용하기

a = [1,2,3,4]
result = [num * 3 for num in a]

print(result)
````

````python
# 조건을 넣어 list comprehension 사용하기

result = [num * 3 for num in a if num % 2 == 0]

print(result)
````

````python
# list comprehension의 중첩 반복문

result = [num * num2 for num in range(1, 5)
                     for num2 in range(1, 3)]
print(result)
````

## 반복문과 조건문의 혼합



````python
# 커피 자판기 만들기

coffee = 3

while True:
  menu = input('''
  메뉴를 골라 주세요
  1. 아메리카노
  2. 라떼
  3. 모카
  ''')
  if menu == "1":
    coffee = coffee -1
    print("아메리카노 나왔습니다!")
    print(f"남은 커피의 양은 {coffee}개입니다.")
  elif menu == "2":
    coffee = coffee -1
    print("라떼 나왔습니다!")
    print(f"남은 커피의 양은 {coffee}개입니다.")
  elif menu == "3":
    coffee = coffee -1
    print("모카 나왔습니다!")
    print(f"남은 커피의 양은 {coffee}개입니다.")
  else:
    print("메뉴 중에서 골라주세요")
  if coffee == 0:
    print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
    break
````

````python
# 더 복잡하게

menu = {"아메리카노":2500, "라떼":3500, "모카":3000}
cash = 0
coffee = 3
flag = True

while coffee > 0:

  # 응답 받고 정합성 검사 파트를 분리
  while True:
    order = input('''
    메뉴를 골라 주세요(숫자로 입력해주세요)
    1. 아메리카노
    2. 라떼
    3. 모카
    4. 그만둔다
    ''')
    if order == "1" or order == "2" or order == "3":
      break
    elif order == "4":
      flag = False
      break
    else:
      print("메뉴 중에서 골라주세요")
      pass
  # flag 변수를 이용해 그만둔다를 선택했을 경우 전체 루프에서 탈출
  if not flag: 
    break

  # order를 (int(order)-1) 하면 menu의 index로 사용 가능
  ind = int(order)-1
  coffee = coffee - 1
  print(f"{list(menu.keys())[ind]} 나왔습니다!")
  cash += menu[list(menu.keys())[ind]]
  print(f"남은 커피의 양은 {coffee}개입니다.")

# 루프가 종료된 뒤 매출 출력
print("금일 영업이 종료되었습니다.") 
print(f"오늘 매출은 {cash}원입니다.")
````

## 반복문으로 여러 연산 해보기

````python
# 랜덤 리스트 생성

import random

list_test = [random.randint(0, 100) for i in range(20)]

list_test
````

````python
# sum 함수

sum(list_test)
````

````python
# 반복문으로 sum 해보기
sum_list = 0
for i in list_test:
  sum_list += i
print(sum_list)
````

````python
# 왜 굳이 반복문으로 sum을?
sum_odd = 0
sum_even = 0
for i in list_test:
  if i % 2 == 1:
    sum_odd += i
  elif i % 2 == 0:
    sum_even += i
print(sum_odd, sum_even, sum_odd + sum_even)
````

````python
# list comprehension 활용해보기
print(sum([i for i in list_test if i % 2 == 0]))
print(sum([i for i in list_test if i % 2 == 1]))
````

````python
# for문 얹기
for j in range(4):
  print(sum([i for i in list_test if i % 4 == j]))
````

````python
# 반복문으로 1~3번째로 큰 수 찾기

max, max2, max3 = 0, 0, 0

for i in list_test:
  if i>max:
    max, max2, max3 = i, max, max2
  elif i>max2:
    max2, max3 = i, max2
  elif i>max3:
    max3 = i
  
print(list_test)
print(max, max2, max3)
````
