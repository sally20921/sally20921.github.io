
````python
text_in = input("두 수를 입력하세요.")
text_in_ = text_in.split()

num_1 = int(text_in_[0])
num_2 = int(text_in_[1])
divisor = []
common_divisor = []
 
# 약수 구하기
for i in range(num_1):
    div = i+1
    if num_1%div == 0:
        divisor.append(div)

# 공약수 구하기
for div in divisor:
    if num_2%div == 0:
         common_divisor.append(div)

# 최대공약수 수하기
gcd = common_divisor[-1]
print(gcd)
````

````python
# 함수 구현
def _GCD(num_1, num_2):
    # 약수 구하기
    for i in range(num_1):
        div = i+1
        if num_1%div == 0:
            divisor.append(div)

    # 공약수 구하기
    for div in divisor:
        if num_2%div == 0:
             common_divisor.append(div)
    # 최대공약수 수하기
    gcd = common_divisor[-1]
    return gcd

# main 함수(아래 코드는 편집x)
def main():
  text_in = input("두 수를 입력하세요.")
  text_in_ = text_in.split()

  num_1 = int(text_in_[0])
  num_2 = int(text_in_[1])

  gcd = _GCD(num_1, num_2)
  print(gcd)

# main 함수 실행
main()
````

#### 같이 해보기 3. swap( ) 함수는 두 매개변수에 할당된 값을 서로 바꿔서 다시 할당하는 기능을 합니다. 아래의 코드를 참고하여 swap( ) 함수를 구현해봅시다.

````python
# x의 값과 y의 값을 swap할 수 있는 코드 예시
x = 10
y = 3

temp = x
x = y
y = temp
print("x: {0}, y: {1} ".format(x, y))
````

````python
def swap(x, y):
  temp = x
  x = y
  y = temp

# main 함수(아래 코드는 편집x)
def main():
  num1 = 10
  num2 = 3
  swap(num1, num2)
  print("num1: {0}, num2: {1} ".format(num1, num2))

# main 함수 실행
main()
````

* 지역변수(Local variable)와 전역변수(Global Variable) 개념 이해
* 할당에 의한 호출(Call by assignment)와 참조에 의한 호출(Call by reference) 개념 이해 </br>
: 함수가 실행되는 과정에서 매개변수가 입력되는 방식에 관한 이해(판서 참고)
  + 정수형, 실수형 변수(int, float), 문자열 변수(str)과 같이 immutable한 자료형의 매개변수는 함수가 실행되는 과정에서 할당에 의해 호출.
  + 리스트, 딕셔너리와 같은 mutable한 자료형의 매개변수는 참조에 의해 호출.
  + mutable과 immutable 개념 구분은 6주차 클래스에서 자세히 다룰 예정.

````python
# '전역변수'를 활용한 방법
num1 = 10
num2 = 3

# swap
def swap():
  global num1 
  global num2
  temp = num1
  num1 = num2
  num2 = temp

# main 함수
def main():
  swap()
  print("num1: {0}, num2: {1} ".format(num1, num2))

# main 함수 실행
main()

````

````python
# '참조에 의한 호출'을 활용한 방법
def swap(_nums):
  temp = _nums[0]
  _nums[0] = _nums[1]
  _nums[1] = temp 

# main 함수(아래 코드는 편집x)
def main():
  num1 = 10 # 지역 변수
  num2 = 3

  nums = [] 
  nums.append(num1)
  nums.append(num2)
  
  swap(nums) # 리스트를 함수에 넘기기

  num1 = nums[0]
  num2 = nums[1]
  print("num1: {0}, num2: {1} ".format(num1, num2))

# main 함수 실행
main()
````

#### Recursive implementation of Fibonacci number

$F_0 = 0$

$F_1 = 1$

$F_n = F_{n-1} + F_{n-2}$

````python
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

fib(5)

# fib(5) # 0, 1, 1, 2, 3, 5
````

````python
def fib(num, flag):    
    if num == 0:
        print("fib({0}), depth: {1}". format(num, flag))
        flag = flag + 1
        return 0
    elif num == 1:
        print("fib({0}), depth: {1}". format(num, flag))
        flag = flag + 1
        return 1
    else:
        print("fib({0}), depth: {1}". format(num, flag))
        flag = flag + 1
        return fib(num-1, flag) + fib(num-2, flag)

fib(4, 1)

# fib(4) # 0, 1, 1, 2, 3
````

#### 메모이제이션(Menoization) 기법을 통해 재귀함수의 실행 속도를 높이기
* 재귀함수의 치명적인 단점은 매개변수의 크기가 증가할수록 프로그램이 수행하는데 드는 시간이 기하급수적(n^2)으로 증가한다. 
* 메모이제이션 기법은 기하급수적으로 증가하는 비용을 선형적(n)으로 증가할 수 있도록 효율성을 높인다,

같이 해보기 7. 아래 주어진 피보나치 수열의 깊이를 출력해봅시다.

````python
# 기억을 하기 위한 메모 공간
fib_dic = {} 
# dictionary로 구현. 이름이 헷갈려서 fib_list에서 fib_dic으로 수정했습니다. 

def fib(n):
    if n in fib_dic:
      return fib_dic[n]
    if n==0:
        fib_dic[0] = 0
        return 0
    elif n==1:
        fib_dic[1] = 1
        return 1
    else:
        fib_dic[n] = fib(n-1) + fib(n-2)
        return fib_dic[n]

# main() 함수
def main():
  fib(4)
  for i in range(5):
    print(fib_dic[i], end = " ")

# main() 함수
main()
````

#### 직접 해보기 2-1. 3항 점화식이 주어져 있습니다. 재귀함수를 이용하여 n이 10일 때 값을 출력해봅시다.

* 3항 점화식

$S_0 = 0$

$S_1 = 1$

$S_2 = 2$

$S_n = 3S_{n-1} + 2S_{n-2} + S_{n-3}$

입출력 예시:
```
정수를 입력하세요.5
105
```

````python
def func(n):
  if n==0:
    return 0
  if n==1:
    return 1
  if n==2:
    return 2
  else:
    return 3*func(n-1)+2*func(n-2)+func(n-3)

# main 함수(아래 코드는 편집x)
def main():
  text_in = input("정수를 입력하세요.")
  num = int(text_in)
  result = func(num)
  print(result)

# main 함수 실행
main()
````
